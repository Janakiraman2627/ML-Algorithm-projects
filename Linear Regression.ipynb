```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/Users/janakiramanprakash/Desktop/netzwerk/assignments/Salary_Data.csv')
```

```python
df.head()
```

```python
x = df['YearsExperience'].values
y = df['Salary'].values
```

```python
mean_x = np.mean(x)
mean_y = np.mean(y)

print(mean_x)
print(mean_y)
```

```python
numerator = 0
denominator = 0

for i in range (len(x)):
    numerator += (x[i] - mean_x)*(y[i]- mean_y)
    denominator += (x[i] - mean_x)**2
    m = numerator/denominator
    c = mean_y - (mean_x*m)
print(m,c)
```

```python
max_x =np.max(x)
min_x = np.min(x)
X = np.linspace(min_x,max_x,50)
Y = m*X + c
plt.plot(X,Y,color = 'g', label = 'linear line' )
plt.scatter(x,y,color = 'r', label = 'data points' )
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.legend()
plt.show()
```

```python
num = 0
den = 0


for i in range (len(x)):
    y_pred = m*x[i]+c
    num += (y_pred - mean_y)**2
    den += (y[i] - mean_y)**2
    
r2_score = num/den
    
```

```python
print(r2_score)
```
